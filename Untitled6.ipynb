### Generating and ARMA process

```r
set.seed(74)
model = arima.sim(list(order=c(1, 0, 2), ar=c(0.8),ma=c(-0.7, 0.1)), sd=1, n=200)
```

```r
plot(model, type='o')
```

```r
acf(model)
```

```r
pacf(model)
```
